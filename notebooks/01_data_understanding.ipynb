# Data Preprocessing
#### Ddealing with missing Values

```python
df.isna().sum()
```

```python
df[["WITHDRAWALAMT", "DEPOSITAMT"]].head(20)
```

```python
print("Depo",df["DEPOSITAMT"].isna().sum())
print("with",df["WITHDRAWALAMT"].isna().sum())
```

```python
# Change the dtype to numeric
df["DEPOSITAMT"] = df["DEPOSITAMT"].astype(str).str.replace(",", "")
df["WITHDRAWALAMT"] = df["WITHDRAWALAMT"].astype(str).str.replace(",", "")
df["DEPOSITAMT"] = pd.to_numeric(df["DEPOSITAMT"], errors="coerce")
df["WITHDRAWALAMT"] = pd.to_numeric(df["WITHDRAWALAMT"], errors="coerce")
```

```python
df["WITHDRAWALAMT"].dtype
```

```python
# When we have a withdrawal, we don't have a deposit and vice versa. So we can fill the missing values with 0.
df["DEPOSITAMT"] =  df["DEPOSITAMT"].fillna(0)
df["WITHDRAWALAMT"] =  df["WITHDRAWALAMT"].fillna(0)
```

```python
df[["WITHDRAWALAMT", "DEPOSITAMT"]].head(10)
```

#### Convert Date Column to Date Time Type

```python
df["VALUEDATE"]= pd.to_datetime(df["VALUEDATE"])
df["TRANSDATE"]= pd.to_datetime(df["TRANSDATE"])
```

```python
df.dtypes
```

#### Creating Customer Level Aggregation w.r.t Transactions

```python
customer_transactions = df.groupby("ACCOUNTNO").agg(
    total_deposit = ("DEPOSITAMT", "sum"),
    total_withdrawal = ("WITHDRAWALAMT", "sum"),
    average_deposit = ("DEPOSITAMT", "mean"),
    average_withdrawal = ("WITHDRAWALAMT", "mean"),
    transaction_count = ("ACCOUNTNO", "count")
).reset_index()



customer_transactions["net_flow"]= (

    customer_transactions["total_deposit"] - customer_transactions["total_withdrawal"]
)
```

#### Creating Customer Attribute Table 

```python
customer_profile =df.groupby("ACCOUNTNO").first().reset_index()
```

```python
customer_profile = customer_profile.drop(
    ["TRANSDATE", "VALUEDATE", "TRANSACTIONDETAILS",'WITHDRAWALAMT', 'DEPOSITAMT'],
    axis=1
)
```

#### Creating Customer data frame

```python
# merging the 2 tables on ACCOUNTNO

customer_df = customer_profile.merge(
    customer_transactions,
     on="ACCOUNTNO",
      how="left"
)
```
